#### Getting geographicals for the district

```python
district_names = ['Mitte', 'Pankow', 'Charlottenburg-Wilmersdorf']

geolocator = Nominatim(user_agent="Berlin_Rent_Analyzer")

rows = []
for district_name in district_names:
    location = geolocator.geocode(district_name + ', Berlin, Germany')
    if location is not None:
        rows.append({'District': district_name,
                     'Latitude': location.latitude,
                     'Longitude': location.longitude})

df = pd.DataFrame(rows)
```
